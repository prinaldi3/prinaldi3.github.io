```python
import pandas as pd

#Opening data in pandas dataframe
df_proteins = pd.read_csv("./Data/Data_Cortex_Nuclear.csv")

#Setting index so all values in table are the same observational unit
df_proteins.set_index(["Genotype", "Treatment", "Behavior", "class", "MouseID"])
```

We downloaded the dataset from Kaggle and it is already in tidy data form. All columns are named properly and are not variables, all empty values are set to NaN, each variable forms a row, all variables have their own column, and there is only one type of observational unit in the table with exception of five columns that are necessary for data categorization (genotype, treatment, behavior, class, and mouseID) which we set as the index for the dataframe. Otherwise, we had no challenges formatting the data since Kaggle had already done a very nice job making sure it was tidy. Downloading the data was also a very easy process and required no web scraping or tricky extraction. 


Looking at the dataframe, there are 15 measurements for each mouse. For our future analysis we will most likely
be taking the mean of the 15 measurements and using that value in our analysis for each protein. There are 77
