## Milestone 2
### Exploratory Analysis

As stated above, we will begin by taking the mean of the 15 measurements of protein expression for each mouse. First, however, a bit more manipulation of the data is necessary. The MouseID column contains two bits of information -- the mouse that was experimented on, and the number of the measurement that any particular row represents (i.e. 309_1 denotes the first measurement taken on mouse 309). We split this column into two columns -- one for the mouse and one for the measurement number -- so that we could then easily group by mouse and take the average of the protein expression data over the 15 trials. 


```python
df_proteins["Mouse"] = df_proteins["MouseID"].str.split("_", expand=True)[0]
df_proteins["Measurement#"] = df_proteins["MouseID"].str.split("_", expand=True)[1]
df_proteins.drop(columns="MouseID", inplace=True)
df_proteins = df_proteins.set_index(["Genotype", "Treatment", "Behavior", "Measurement#"])
df_proteins
```

```python
mean_expressions = df_proteins.groupby("Mouse").mean()
mean_expressions
```

There are many experimental variables in this data set, and we thought it would be a good idea to first get a sense of how protein expression varies when controlling for a few of these variables. We will begin by examining the differences in protein expression for the control mice who were stimulated to learn when given either saline or memantine. The two classes of interest are thus c-CS-s and c-CS-m.

```python
control_stim_saline = df_proteins[df_proteins["class"] == "c-CS-s"]
control_stim_mem = df_proteins[df_proteins["class"] == "c-CS-m"]
```

Now, we just want the mice in the c-CS-s and c-CS-m classes, since this will allow us to index into the mean_expressions DataFrame and retrieve only the relevant data.

Now we're going to analyze some of the proteins. A useful piece of analysis would be to find out which proteins have similar expression levels to determine if the proteins are possibly expressed through the same mechanism. If several proteins do have the same or similar expression mechanisms, looking at the differences in variables such as treated with saline or memantine could give us a clue what proteins or pathways memantine affects in the brain, or another variable such as being stimulated to learn. Let's take the protein pCREB_N and analyze the protein that has the most similar expression to it for mice stimulated to learn and either treated with memantine or saline. 

```python
from sklearn.metrics import pairwise_distances

# Making table for all the proteins for mice treated with saline and stimulated to learn
saline_expressionsA = mean_expressions.loc[saline_mice][df_proteins.columns[:len(df_proteins.columns) - 2]]

# Making table for all the proteins for mice treated with memantine and stimulated to learn
memantine_expressionsA = mean_expressions.loc[memantine_mice][df_proteins.columns[:len(df_proteins.columns) - 2]]

# Getting the mean for each protein expression for all mice in the same class
salineProteins = saline_expressionsA.mean()
memantineProteins = memantine_expressionsA.mean()

sdf = salineProteins.to_frame().reset_index()


# Getting index of pCREB_N
pCREB_N  = int(sdf[sdf["index"]=="pCREB_N"].index.values)




# Calculating euclidean distance and finding the cloest neighbor for saline treated mice
salineProteins = salineProteins.to_frame(name = "Average Expression")
D = pairwise_distances(salineProteins, metric="euclidean")
D.shape
np.fill_diagonal(D, np.inf)
D[pCREB_N, :].argmin()
# index of closest protein is 5

salineProteins.iloc[5]
```

```python
saline_expressionsA
```

```python
# Calculating euclidean distance and finding the cloest neighbor for memantine treated mice
memantineProteins = memantineProteins.to_frame(name = "Average Expression")
D = pairwise_distances(memantineProteins, metric="euclidean")
D.shape
np.fill_diagonal(D, np.inf)
D[pCREB_N, :].argmin()

# index of closest protein is 56
memantineProteins.iloc[56]
```

For mice treated with memantine (c-CS-m), we have found that the protein expression closest to pCREB_N is pAKT_N. For mice treated with saline (c-CS-s), the closest protein expression to pCREB_N is Tau_N. It is interesting that pCREB_N expression is closest to a different protein for each class. Let's further look at this by graphing these three proteins and their expressions for both classes of mice. 

```python
# Saline Mice
sal3 = salineProteins.iloc[[5, pCREB_N, 56], :].reset_index()

# Memantine Mice
mem3 = memantineProteins.iloc[[5, pCREB_N, 56], :].reset_index()

# Making Dataframe and Graph
df = pd.merge(sal3, mem3, on='index')
df = df.rename(columns = {"index":"Protein", "Average Expression_x":"c-CS-s", "Average Expression_y": "c-CS-m"})
df = df.set_index("Protein")
ax = df.plot(kind='barh', title = "Average Protein Expression")
plt.xlabel("Protein Expression Level")

```

The bar graph above shows the expression level of a certain protein in order to directly compare the two sets of mice, those treated with saline and those treated with memantine. As we can see from this graph there isn't a huge change in expression of any of the proteins between the two sets of mice which means these proteins do not have any sort of correlation. We do have a total of 77 proteins and four different variables to compare so this will be a useful visual when looking for proteins that might have a correlation.
